### Data Analysis and Visualization

```python
state_orders = data.groupby('State')['Orders'].sum()
```

```python
state_orders.plot(kind='bar')

plt.title('State-wise Total Orders')
plt.xlabel('State')
plt.ylabel('Total Orders')
#ax.bar_label(ax.containers[0])(bar ke uper value dikhani hai to)
plt.show()
```

```python
# the highest number of orders were placed in utter Pardesh and lowest telangana
```

```python
zone_orders = data.groupby('Zone')['Orders'].sum()
```

```python
zone_orders.plot(kind='pie', autopct='%1.1f%%')

plt.title('Zone Wise Orders Distribution')

plt.ylabel('')  # Y-axis hata deta hai
plt.show()
```

```python
# the highest number of orders were zone in central and lowest eastern
```

```python
catgory_orders = data.groupby('Product_Category')['Orders'].sum()
```

```python
catgory_orders.plot(kind='bar',width = 0.8)

plt.title('Product_Category-wise Total orders'  )
plt.xlabel('Product_Category')
plt.ylabel('total_Orders')
plt.show()
```

```python
#  the clothing & apparel catgory has the highest number of orders and lowest hand & power tools
```

```python
Occupation_orders = data.groupby('Occupation')['Orders'].sum()
```

```python
Occupation_orders.plot(kind='bar',width = 0.8)

plt.title('Occupation-wise total orders')
plt.xlabel('Occupation')
plt.ylabel('Orders')
plt.show()
```

```python
#  the IT sector Occupation has the highest number of orders and lowest agriculture
```

```python
state_amount = data.groupby('State')['Amount'].sum()
```

```python
state_amount.plot(kind='bar',width = 0.8)

plt.title('state-wise amount')
plt.xlabel('State')
plt.ylabel('total_amount')

plt.show()
```

```python
# highest amount in utter pardesh and lowest amount telangana
```

```python
age_group_amount = data.groupby('Age Group')['Amount'].sum()
```

```python
age_group_amount.plot(kind='bar',width = 0.8)

plt.title('age group-wise total amount')
plt.xlabel('Age Group')
plt.ylabel('total_amount')

plt.show()
```

```python
# highest amount in age group 26-35 and lowest amount in age group 0-17 
```

```python
gender_orders = data.groupby('Gender')['Orders'].sum()
```

```python
gender_orders.plot(kind='pie', autopct='%1.1f%%')

plt.title('gender Wise order Distribution')

plt.ylabel('')  # Y-axis hata deta hai
plt.show()
```

```python
ax = sns.countplot(x = 'Gender',data = data)
ax.bar_label(ax.containers[0])
```

```python
ax = sns.countplot(x = 'Age Group',data = data,hue = 'Gender')
ax.bar_label(ax.containers[0])
```
